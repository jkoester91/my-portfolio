# Step 5 : Hypothesis Tests

‚ÄîAverage user ratings of the Xbox One and PC platforms are the same.

‚ÄîAverage user ratings for the Action and Sports genres are different.

Set the *alpha* threshold value yourself.

Explain:

```python
# Hypothesis 1: Average user ratings of the Xbox One and PC platforms are the same.
xbox1 = df_relevant[df_relevant['platform'] == 'XOne']['user_score']
xbox1 = pd.to_numeric(xbox1, errors='coerce')
xbox1 = xbox1.dropna()

pc = df_relevant[df_relevant['platform'] == 'PC']['user_score']
pc = pd.to_numeric(pc, errors='coerce')
pc = pc.dropna()

# calculate the p-value
alpha = 0.05
plat_results = st.ttest_ind(xbox1,pc)

print('p-value: ', plat_results.pvalue)

if plat_results.pvalue < alpha:
    print("We reject null hypothesis: Average user ratings for Xbox One and PC are different.")
else:
    print("We can't reject the null hypothesis: Average user ratings for Xbox One and PC are equal.")
```

```python
# Hypothesis 2: Average user ratings for the Action and Sports genres are different. 
action = df_relevant[df_relevant['genre'] == 'Action']['user_score']
action = pd.to_numeric(action, errors='coerce')
action = action.dropna()

sports = df_relevant[df_relevant['genre'] == 'Sports']['user_score']
sports = pd.to_numeric(sports, errors='coerce')
sports = sports.dropna()

# set p-value
alpha = 0.05
genre_results = st.ttest_ind(action,sports)

print('p-value: ', genre_results.pvalue)

if genre_results.pvalue < 0.05:
    print('We reject the null hypothesis: Average user ratings for Action and Sports genres are different.')
else:
    print("We can't reject the null hypothesis: Average user ratings for Action and Sports genres are equal.")
```
